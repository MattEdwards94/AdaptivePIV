# Interpolation
This notebook intends to clarify some of the confusion around interpolation, polynomials, and splines.

## Polynomial
### degree
The `degree` refers to the largest sum of exponents in the equation.
For example `ax^3 + bx^2 + cx + d` is a cubic polynomial with degree 3, since the largest power is 3.
Conversely `ax^3y^2 + bx^2 + cx + d` would be a two dimensional polynomial with degree 5

### Cubic polynomial
Given the value of a function at two points `x1` and `x2`, as well as the function derivative at these two points, there exists only one cubic function which can satisfy these conditions.


### Piecewise
"A piecewise-defined function is a function which is deinfed by multiple sub-functions, with each sub-function applying to a certain interval of the main functions domain" - https://en.wikipedia.org/wiki/Piecewise

The term piecewise can also be used to define the properties of each piece, but which may not apply to the whole function. 
For example, a function is piecewise differentiable if each piece is differentiable over the pieces subdomain, even though the whole function may not be differentiable.


### Continuity
*  C<sup>0</sup> - curve is continuous
*  C<sup>1</sup> - first derivative is continuous
*  C<sup>2</sup> - second derivative is continuous


### Hermite
A hermite interpolation matches an unknown function both in the observed value, and in the observed value of the its first $m$ derivatives.
I am not sure exactly what Hermite form is.

## Spline
A spline is a piecewise polynomial 


### Cubic spline / cubic hermite spline
This is a spline where each piece of the piecewise polynomial is a third degree polynomial specified in 'hermite form'.
The method requires the values and the derivatives at each end of the interval $(x_k, x_k+1)$ - if the derivatives are not available then they must be estimated.

The resulting spline is C<sup>1</sup> continuous.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as sint
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20,10)
```

Check to see the support kernel of various functions in 1D

```python
# define the signal function
def ft(x):
    return np.sin(x*x)
```

```python
# plot the function over a range of values
xt = np.linspace(0, 2*np.pi, 500)
plt.plot(xt, ft(xt))
```

```python
# create function to perform several interpolations at once
def interp(xs, ys):
    # interpolate linearly
    fint_lin = sint.interp1d(xs, ys, kind='linear')
    # interpolate using a first order spline
    fint_slin = sint.interp1d(xs, ys, kind='slinear')
    # interpolate using a third order spline
    #fint_cub = sint.interp1d(xs, ys, kind='cubic')
    fint_cub = sint.CubicSpline(xs, ys, bc_type='clamped')
    fint_cubk = sint.CubicSpline(xs, ys, bc_type='not-a-knot')
    fint_akima = sint.Akima1DInterpolator(xs, ys)
    # interpolate using PCHIP
    fint_pchip = sint.PchipInterpolator(xs, ys)
    
    
    return fint_lin, fint_slin, fint_cub, fint_pchip, fint_cubk, fint_akima
```

```python
def plot(ax, xs, ys, xeval, interpFunc, label_in, colour, shape, style):
    ax.plot(xeval, interpFunc(xeval), color=colour, linestyle=style, label=label_in)
    ax.plot(xs, ys, color=colour, marker=shape, linestyle='')
    
```

```python
# create a number of sample points
xs = np.linspace(0, 2*np.pi, 30)
ys = ft(xs)

# interpolate using different interpolants
flin, fslin, fcub, fpchip, fcubk, fakima = interp(xs, ys)

# plot the results
fig, ax = plt.subplots()
ax.plot(xt, ft(xt), label='Signal')
#plot(ax, xs, ys, xt, flin, 'Linear', 'g', 'x', '-')
plot(ax, xs, ys, xt, fcub, 'Cubic', 'm', 'x', '-')
plot(ax, xs, ys, xt, fcubk, 'Cubic', 'k', 'x', '-')
plot(ax, xs, ys, xt, fakima, 'Akima', 'g', 'x', '-')
#plot(ax, xs, ys, xt, fpchip, 'Pchip', 'k', 'x', '-')
ax.legend()
np.allclose(fcubk(xt), fcub(xt))
```

## Perturbing points


```python
ys[1] = ft(xs[1])*1.5  # peturb the value of a sample slightly

flinp, fslinp, fcubp, fpchipp, fcubkp, fakimap= interp(xs, ys)

fig, ax = plt.subplots()
#ax.plot(xt, ft(xt), label='Signal')
ax.plot(xs, np.zeros(len(xs)), 'x');
ax.plot(xt, fcubp(xt)-fcub(xt), 'm-', label='Cubic spline')
ax.plot(xt, fcubkp(xt)-fcubk(xt), 'k-', label='Cubic spline - not a knot')
ax.plot(xt, fakimap(xt)-fakima(xt), 'g-', label='Akima')

#ax.plot(xt, fpchipp(xt)-fpchip(xt), 'k-', label='PChip')
ax.set_ylim([-0.000005, 0.000005])
ax.legend()

print(fcubp(xs[:-1:]+0.01)-fcub(xs[:-1:]+0.01))
print(fcubkp(xs[:-1:]+0.01)-fcubk(xs[:-1:]+0.01))
print(fakimap(xs[:-1:]+0.01)-fakima(xs[:-1:]+0.01))
```
